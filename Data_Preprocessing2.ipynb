```python
import numpy as np
import pandas as pd
import re
import emoji

df = pd.read_csv('davidson.csv')
```

```python
df = df[['tweet', 'class']]

df = df.rename(columns={'tweet': 'text', 'class': 'label'})


label_map = {0: 'hate', 1: 'offensive', 2: 'normal'}
df['label'] = df['label'].map(label_map)


def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = emoji.replace_emoji(text, replace='')
    return text


df['clean_text'] = df['text'].apply(clean_text)


df = df[df['clean_text'].str.strip() != '']


df.to_csv("davidson_preprocessed.csv", index=False)

df.head()
```

```python
import json
import pandas as pd
import re, emoji
from collections import Counter


with open("dataset.json", 'r') as f:
    data = json.load(f)

records = []
for post_id, post_data in data.items():
    tokens = post_data.get('post_tokens', [])
    text = " ".join(tokens)


    labels = [ann.get('label') for ann in post_data.get('annotators', []) if 'label' in ann]


    label = Counter(labels).most_common(1)[0][0] if labels else None

    records.append({
        "text": text,
        "label": label
    })

df = pd.DataFrame(records)


df = df.dropna(subset=['label'])


label_map = {
    'hate-speech': 'hate',
    'offensive': 'offensive',
    'normal': 'normal'
}
df['label'] = df['label'].map(label_map)


def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^a-z\s]', '', text)
    text = re.sub(r'\s+', ' ', text).strip()
    text = emoji.replace_emoji(text, replace='')
    return text

df['clean_text'] = df['text'].apply(clean_text)
df = df[df['clean_text'].str.strip() != '']


df = df[['text', 'label', 'clean_text']]



```

```python

import re, emoji

def clean_text(text):
    text = str(text).lower()
    text = re.sub(r'@\w+', '', text)
    text = re.sub(r'http\S+|www\S+', '', text)
    text = re.sub(r'[^a-z\u0900-\u097F\s]', '', text)
    text = emoji.replace_emoji(text, replace='')
    text = re.sub(r'\s+', ' ', text).strip()
    return text




import joblib

model = joblib.load('lang_class_model.pkl')
vectorizer = joblib.load('lang_vectorizer.pkl')
print("Loaded model and vectorizer.")


def predict_language(text):
    clean = clean_text(text)
    vector = vectorizer.transform([clean])
    pred = model.predict(vector)[0]
    return pred



print(predict_language("what is this"))
print(predict_language("aap kaise ho"))
print(predict_language("mai kal office jaunga"))
```
